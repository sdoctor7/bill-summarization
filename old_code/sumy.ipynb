### Sumy - KL and LexRank

```python
train_data = pd.read_pickle('data/train.pkl')
dev_data = pd.read_pickle('data/dev.pkl')

NUM_SENTENCE = 4

summary_kl = train_data['body.section'].apply(run_sumy, algo='KL', sent_count=NUM_SENTENCE)
summary_lexrank = train_data['body.section'].apply(run_sumy, algo='LexRank', sent_count=NUM_SENTENCE)

train_data['summary-kl'] = summary_kl
train_data['summary-lexrank'] = summary_lexrank

train_data.to_pickle('data/sumy.pkl')
```
